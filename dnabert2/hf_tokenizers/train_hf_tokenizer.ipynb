### Unigram Tokenizer with SentencePiece

```python
tokenizer = Tokenizer(models.Unigram())
```

```python
from tokenizers import Regex

tokenizer.normalizer = normalizers.Sequence(
    [
        normalizers.Replace("``", '"'),
        normalizers.Replace("''", '"'),
        normalizers.NFKD(),
        normalizers.StripAccents(),
        normalizers.Replace(Regex(" {2,}"), " "),
    ]
)
```

```python
# SentencePiece Pre-Tokenizer
tokenizer.pre_tokenizer = pre_tokenizers.Metaspace()
```

```python
# no normalization, because not necessary in DNA sequences
```

```python
tokenizer.pre_tokenizer.pre_tokenize_str(sample['text'])
```

```python
# unk token is Unigram Specific
special_tokens = ["<cls>", "<sep>", "<unk>", "<pad>", "<mask>", "<s>", "</s>"]
trainer = trainers.UnigramTrainer(
    vocab_size=25000, special_tokens=special_tokens, unk_token="<unk>"
)
tokenizer.train_from_iterator(get_training_corpus(), trainer=trainer)
```

```python
encoding = tokenizer.encode(sample['text'])
print(encoding.tokens)
```

```python
cls_token_id = tokenizer.token_to_id("<cls>")
sep_token_id = tokenizer.token_to_id("<sep>")
print(cls_token_id, sep_token_id)
```

```python
tokenizer.post_processor = processors.TemplateProcessing(
    single="$A:0 <sep>:0 <cls>:2",
    pair="$A:0 <sep>:0 $B:1 <sep>:1 <cls>:2",
    special_tokens=[("<sep>", sep_token_id), ("<cls>", cls_token_id)],
)
```

```python
encoding = tokenizer.encode(sample['text'])
print(encoding.tokens)
print(encoding.type_ids)
```

# DNABERT2 Tokenizer Code:

```python
from tokenizers import Tokenizer
from tokenizers.models import BPE
from tokenizers.trainers import BpeTrainer
from tokenizers.pre_tokenizers import Whitespace
from tokenizers.processors import TemplateProcessing
from datasets import load_dataset

from tokenizers import (
    decoders,
    models,
    normalizers,
    pre_tokenizers,
    processors,
    trainers,
    Tokenizer,
)
```

```python
dataset_dir = '/scratch/jbusch/ma/data/dnabert_2_pretrain/'
train_file_name = 'train_sub5m.txt'
vocab_size = 4096
special_tokens = ["[UNK]", "[CLS]", "[SEP]", "[PAD]", "[MASK]"]
```

```python
# tokenizer = Tokenizer(models.BPE())
# tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
# trainer = trainers.BpeTrainer(vocab_size=4096, special_tokens=["<|endoftext|>"])
# tokenizer.train_from_iterator(get_training_corpus(), trainer=trainer)
```

```python
tokenizer = Tokenizer(BPE(unk_token=special_tokens[0]))
trainer = BpeTrainer(special_tokens=special_tokens, vocab_size = vocab_size, min_frequence=2)
```

```python
tokenizer.pre_tokenizer = Whitespace()
# tokenizer.pre_tokenizer = pre_tokenizers.ByteLevel(add_prefix_space=False)
```

```python
trainer.special_tokens
```

```python
train_data = load_dataset('text', data_files={'train': f'{dataset_dir}/{train_file_name}',
                                              'val': f'{dataset_dir}/{valid_file_name}'})
dataset = train_data["val"]
def get_training_corpus():
    for i in range(0, len(dataset), 1000):
        yield dataset[i : i + 1000]["text"]
```

```python
tokenizer.train_from_iterator(get_training_corpus(), trainer=trainer)
```

```python
tokenizer.post_processor = TemplateProcessing(
    single="[CLS] $A [SEP]",
    pair="[CLS] $A [SEP] $B:1 [SEP]:1",
    special_tokens=[
        ("[CLS]", tokenizer.token_to_id("[CLS]")),
        ("[SEP]", tokenizer.token_to_id("[SEP]")),
    ],
)
```

```python
tokenizer.save(f"/scratch/jbusch/ma/tokenizer/tokenizer_subsample_5M_val.json")
```

```python
tokenizer.train([f'{dataset_dir}{train_file_name}'], trainer)
```
